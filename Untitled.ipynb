```python
import re
import csv

# Apresentação do trabalho
while True:
    seu_nome = input('Olá, qual é o seu nome? ')
    if len(seu_nome) > 40:
        print(
            'ERRO: O nome não pode ter mais do que 40 caracteres. Por favor, insira um nome mais curto.'
        )
    elif len(seu_nome) == 0:
        print(
            'ERRO: O nome não pode estar em branco. Por favor, insira um nome válido.'
        )
    elif seu_nome.isdigit():
        print(
            'ERRO: O nome não pode ser um número. Por favor, insira um nome válido.'
        )
    elif not re.match("^[A-Za-z ]+$", seu_nome):
        print(
            'ERRO: O nome não pode conter símbolos e acentos. Por favor, insira um nome válido.'
        )
    else:
        print(
            f'Ok, {seu_nome}! Estou fazendo seu cadastro e preciso de alguns dados.'
        )
        break

# Inicializando uma lista vazia para armazenar os dados
dados = []

# Carregando dados existentes de um arquivo CSV
try:
    with open("dados.csv", "r") as arquivo_csv:
        leitor_csv = csv.DictReader(arquivo_csv)
        for linha in leitor_csv:
            dados.append(linha)
    print("Dados carregados com sucesso.")
except FileNotFoundError:
    print("Arquivo não encontrado. Criando novo arquivo.")


# Função para criar um novo registro
def criar_registro():
    while True:
        nome = input("Digite o nome: ")
        if re.match(r"^[a-zA-Z\s]+$", nome):
            break
        else:
            print("Por favor, insira um nome válido contendo apenas letras.")
    while True:
        try:
            idade = int(input("Digite a idade: "))
            if idade < 18:
                print(
                    'A idade deve ser no mínimo 18 anos. Por favor, insira uma idade válida.'
                )
            elif idade > 120:
                print(
                    'A idade não pode ser superior a 120 anos. Por favor insira uma idade válida.'
                )
            else:
                break
        except ValueError:
            print("Por favor, insira um número inteiro válido para a idade.")
    while True:
        email = input("Digite o e-mail: ")
        if not re.match(r"[^@]+@[^@]+\.[^@]+", email):
             print("ERRO: O e-mail deve conter o símbolo '@'. Por favor, insira um e-mail válido.")
        else:
            break
    CPFs_cadastrados = [registro["CPF"] for registro in dados]
    while True:
        try:
            CPF = int(input("Digite seu CPF (11 dígitos): "))
            if CPF in CPFs_cadastrados:
                print("CPF já cadastrado. Por favor, insira um CPF diferente.")
            elif len(str(CPF)) != 11:
                print(
                    "O CPF deve ter 11 dígitos. Por favor, insira um CPF válido."
                )
            else:
                break
        except ValueError:
            print("Por favor, insira um número inteiro válido para o CPF.")
    while True:
        try:
            celular = int(input("Digite seu celular: "))
            break
        except ValueError:
            print(
                "Por favor, insira um número inteiro válido para o número do celular."
            )
    dados.append({
        "nome": nome,
        "idade": idade,
        "email": email,
        "CPF": CPF,
        "Celular": celular
    })
    print("Registro criado com sucesso!")


# Função para ler um registro existente


# Função para atualizar um registro existente
# Função para atualizar um registro existente
def atualizar_registro():
    cpf = input("Digite o CPF a ser atualizado: ")
    for registro in dados:
        if registro["CPF"] == cpf:
            while True:
                nome = input('Digite o novo nome: ')
                if re.match(r"^[A-Za-z\s]+$", nome):
                    registro["nome"] = nome
                    break
                elif len(seu_nome) == 0:
                    print(
                        'ERRO: O nome não pode estar em branco. Por favor, insira um nome válido.'
                    )
                elif seu_nome.isdigit():
                    print(
                        'ERRO: O nome não pode ser um número. Por favor, insira um nome válido.'
                    )
                elif not re.match("^[A-Za-z ]+$", seu_nome):
                    print(
                        'ERRO: O nome não pode conter símbolos e acentos. Por favor, insira um nome válido.'
                    )
                else:
                    print(
                        'Por favor, insira um nome válido contendo apenas letras.'
                    )
            while True:
                try:
                    idade = int(input("Digite a idade: "))
                    if idade < 18:
                        print(
                            'A idade deve ser no mínimo 18 anos. Por favor, insira uma idade válida.'
                        )
                    elif idade > 120:
                        print(
                            'A idade não pode ser superior a 120 anos. Por favor insira uma idade válida.'
                        )
                    else:
                        break
                except ValueError:
                    print(
                        "Por favor, insira um número inteiro válido para a idade."
                    )
            while True:
                email = input("Digite o novo e-mail ou existente: ")
                if not re.match(r"[^@]+@[^@]+\.[^@]+", email):
                    print(
                        "ERRO: O e-mail deve conter o símbolo '@'. Por favor, insira um e-mail válido."
                    )
                else:
                    break
            registro["Celular"] = int(
                input("Digite o novo número do celular ou existente: "))
            print("Registro atualizado com sucesso!")
            return
    print("Registro não encontrado.")


# Função para excluir um registro existente
def excluir_registro():
    cpf = input("Digite o CPF a ser excluído: ")
    for registro in dados:
        if registro["CPF"] == cpf:
            dados.remove(registro)
            print("Registro excluído com sucesso!")
            return
    print("Registro não encontrado.")


# Função para ler todos os registros
def ler_registro():
    cpf = input("Digite o CPF a ser buscado: ")
    for registro in dados:
        if registro["CPF"] == cpf:
            print("Nome:", registro["nome"])
            print("Idade:", registro["idade"])
            print("E-mail:", registro["email"])
            print("CPF:", registro["CPF"])
            print("Celular:", registro["Celular"])
            return
    print("Registro não encontrado.")



# Função para salvar os dados em um arquivo CSV
def salvar_em_csv():
    with open("dados.csv", "w", newline="") as arquivo_csv:
        campos = ["nome", "idade", "email", "CPF", "Celular"]
        escritor_csv = csv.DictWriter(arquivo_csv, fieldnames=campos)
        escritor_csv.writeheader()
        escritor_csv.writerows(dados)


while True:
    print()
    print("Selecione uma operação:")
    print("1 - Criar registro")
    print("2 - Atualizar registro")
    print("3 - Excluir registro")
    print("4 - Ler o registro")
    print("5 - salvar dados")
    print("0 - Sair")

    opcao = input("Escolha uma opção: ")

    if opcao == "1":
        criar_registro()
    elif opcao == "2":
        atualizar_registro()
    elif opcao == "3":
        excluir_registro()
    elif opcao == "4":
        ler_registro()
    elif opcao == "5":
        salvar_em_csv()
    elif opcao == "0":
        print(f'Ok! {seu_nome}, volte sempre')
        break
    else:
        print("Opção inválida. Tente novamente.")

# Idade não pode inserir números
# CPF é obrigatoriamente 11 dígitos
# Máximo de caracteres no nome
# Ao apertar enter e deixar campos vazios, mensagem de erro

# Idade não pode inserir números
# CPF é obrigatóriamente 11 dígitos
# Máximo de caracteres no nome
# Ao apertar enter e deixar campos vázios, mensagem de erro
# Limite de caracter na função criar registro NOME
# Colocar @ na função criar registro EMAIL
# Proibir simbolos no nome

# Na operação Atualizar Registro está com os mesmos comandos do Criar Registro
# Idade mínima 18 anos, idade máxima 120 anos.
```
